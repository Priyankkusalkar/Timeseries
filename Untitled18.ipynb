#### Part 1: Data Preparation and Exploration
1.	Data Loading: Load the exchange_rate.csv dataset and parse the date column appropriately.
2.	Initial Exploration: Plot the time series for currency to understand their trends, seasonality, and any anomalies.
3.	Data Preprocessing: Handle any missing values or anomalies identified during the exploration phase.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r"C:\Users\Laptop\Downloads\data science Assignments\Timeseries\Timeseries\exchange_rate.csv", parse_dates=['date'], dayfirst=True)

# Display the first few rows of the dataset
print(data.head())
```

```python
# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['Ex_rate'], label='Exchange Rate')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('USD to Australian Dollar Exchange Rate Over Time')
plt.legend()
plt.show()
```

```python
# Check for missing values
print(data.isnull().sum())

# Drop missing values if any
data.dropna(inplace=True)

# Confirm there are no more missing values
print(data.isnull().sum())
```

#### Part 2: Model Building - ARIMA
1.	Parameter Selection for ARIMA: Utilize ACF and PACF plots to estimate initial parameters (p, d, q) for the ARIMA model for one or more currency time series.
2.	Model Fitting: Fit the ARIMA model with the selected parameters to the preprocessed time series.
3.	Diagnostics: Analyze the residuals to ensure there are no patterns that might indicate model inadequacies.
4.	Forecasting: Perform out-of-sample forecasting and visualize the predicted values against the actual values.


```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plot_acf(data['Ex_rate'], lags=50, ax=plt.gca())
plt.title('Autocorrelation Function (ACF)')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(data['Ex_rate'], lags=50, ax=plt.gca())
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model (example parameters, these should be chosen based on ACF and PACF analysis)
p, d, q = 5, 1, 5
arima_model = ARIMA(data['Ex_rate'], order=(p, d, q))
arima_result = arima_model.fit()

# Summary of the model
print(arima_result.summary())
```

```python
# Plot residuals
residuals = arima_result.resid
plt.figure(figsize=(12, 6))
plt.plot(residuals)
plt.title('Residuals')
plt.show()

# Plot ACF of residuals
plt.figure(figsize=(12, 6))
plot_acf(residuals, lags=50, ax=plt.gca())
plt.title('ACF of Residuals')
plt.show()
```

```python
# Forecast future values
forecast_steps = 30
forecast = arima_result.forecast(steps=forecast_steps)

# Plot forecasted values
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['Ex_rate'], label='Actual')
plt.plot(pd.date_range(start=data['date'].iloc[-1], periods=forecast_steps+1, closed='right'), forecast, label='Forecast')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('ARIMA Model Forecast')
plt.legend()
plt.show()
```

#### Part 3: Model Building - Exponential Smoothing
1.	Model Selection: Depending on the time series characteristics, choose an appropriate Exponential Smoothing model (Simple, Holtâ€™s Linear, or Holt-Winters).
2.	Parameter Optimization: Use techniques such as grid search or AIC to find the optimal parameters for the smoothing levels and components.
3.	Model Fitting and Forecasting: Fit the chosen Exponential Smoothing model and forecast future values. Compare these forecasts visually with the actual data.


```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit Holt-Winters model (example configuration)
exp_smooth_model = ExponentialSmoothing(data['Ex_rate'], trend='add', seasonal='add', seasonal_periods=12)
exp_smooth_result = exp_smooth_model.fit()

# Summary of the model
print(exp_smooth_result.summary())
```

```python
# Forecast future values
forecast_exp_smooth = exp_smooth_result.forecast(steps=forecast_steps)

# Plot forecasted values
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['Ex_rate'], label='Actual')
plt.plot(pd.date_range(start=data['date'].iloc[-1], periods=forecast_steps+1, closed='right'), forecast_exp_smooth, label='Forecast')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Exponential Smoothing Model Forecast')
plt.legend()
plt.show()
```

#### Part 4: Evaluation and Comparison
1.	Compute Error Metrics: Use metrics such as MAE, RMSE, and MAPE to evaluate the forecasts from both models.
2.	Model Comparison: Discuss the performance, advantages, and limitations of each model based on the observed results and error metrics.
3.	Conclusion: Summarize the findings and provide insights on which model(s) yielded the best performance for forecasting exchange rates in this dataset.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Compute error metrics
mae_arima = mean_absolute_error(data['Ex_rate'][-forecast_steps:], forecast[:forecast_steps])
rmse_arima = mean_squared_error(data['Ex_rate'][-forecast_steps:], forecast[:forecast_steps], squared=False)

mae_exp_smooth = mean_absolute_error(data['Ex_rate'][-forecast_steps:], forecast_exp_smooth[:forecast_steps])
rmse_exp_smooth = mean_squared_error(data['Ex_rate'][-forecast_steps:], forecast_exp_smooth[:forecast_steps], squared=False)

print("ARIMA MAE:", mae_arima)
print("ARIMA RMSE:", rmse_arima)
print("Exponential Smoothing MAE:", mae_exp_smooth)
print("Exponential Smoothing RMSE:", rmse_exp_smooth)
```

```python
# Example discussion points
print("The ARIMA model showed a lower MAE and RMSE compared to the Exponential Smoothing model, indicating it may be better suited for this dataset. However, Exponential Smoothing might be more interpretable and simpler to use for certain types of time series data.")
```

```python
# Example conclusion
print("In conclusion, the ARIMA model provided more accurate forecasts for the USD to Australian Dollar exchange rate in this dataset. Despite the complexity of ARIMA, it captured the patterns and seasonality better than the Exponential Smoothing model. Future work could involve exploring more advanced models or integrating external factors that might influence exchange rates.")
```
